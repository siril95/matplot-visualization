```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,10,0.1)
y = 3*x+5

plt.figure(figsize = (5,5))

plt.plot(x,y)
plt.title('plot graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x = np.arange(1,10,0.1)
y = 3*x+5
z = 5*x+4

plt.figure(figsize = (5,5))

plt.subplot(1,2,1)
plt.plot(x,y, color = 'r')
plt.title('plot graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.subplot(1,2,2)
plt.title('plot graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(y,z)
```

```python
# barplot

fruits = {'apple': 30, 'mango': 40, 'orange': 60 }
names  = list(fruits.keys())
quantity = list(fruits.values())
names , quantity

plt.bar(names , quantity, color = ('orange', 'blue', 'green'))
plt.xlabel('names')
plt.ylabel('quantity')
plt.title('distribution of fruits')
plt.show()

```

```python
# scatter plot
x= [10,20,30,40,50]
a = [1,2,3,4,5,]
b = [50,60,70,80, 90]

plt.figure(figsize = (5,5))
plt.scatter(x,a, color = 'black',s=200, marker = '4')
plt.scatter(x,b, color = 'yellow',s= 200, marker = '4')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('scatter plot')
plt.legend(['a', 'b'])
```

```python
# histogram
data = [1,1,1,3,3,6,7,8,4,6,7,3,5,6,7,8,10]
    
plt.hist(data , bins= 30)
plt.show()
```

```python
# boxplot
one = [1,2,3,4,5,6,7,8,9,]
two = [1,3,5,7,9,11,13,15,]
three = [6,7,8,9,10,11,12,12,]
data = list([one,two,three])

plt.boxplot(data)
plt.title('boxplot', color = 'green')
plt.show()
```

```python
#violin plot
one = [1,2,3,4,5,6,7,8,9,]
two = [1,3,5,7,9,11,13,15,]
three = [6,7,8,9,10,11,12,12,]
data = list([one,two,three])

plt.violinplot(data, showmedians  = True,showmeans = True)
plt.title('violin', color = 'green')
plt.show()
```

```python
#pie chart

fruits = {'apple': 30, 'orange': 40, 'mango': 50, 'grapes': 70}
names  = list(fruits.keys())
quantity = list(fruits.values())
plt.title('piechart')
names, quantity

plt.pie(quantity, labels = names, autopct = '%0.1f%%', explode = (0,0,0.1,0.1))
plt.show()
        
```

```python
# area plot

x= [10,20,30,40,50,60,70,80,90]
y = [1,1,3,5,7,9,0,3,5,]

plt.stackplot(x,y)
plt.plot(x,y, color = 'red')
plt.title('area plot')
plt.xlabel('x axis')
plt.ylabel('yaxis')
plt.show()
```
